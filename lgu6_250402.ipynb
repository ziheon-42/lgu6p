# 상위 5개만 보기

- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nlargest.html

- 관련 메서드
+ DataFrame.nsmallest
    Return the first n rows ordered by columns in ascending order.

+ DataFrame.sort_values
    Sort DataFrame by the values.

+ DataFrame.head
    Return the first n rows without re-ordering.

```python
# 숫자열을 sort() 내림차순 정렬
# 상위 5개만 인덱싱
iris.nlargest(5,"sepal_length")
```

```python
iris.nsmallest(5,"sepal_length")
```

# loc vs iloc
- 코드 비교

## loc

```python
#tips.loc[tips['day'] == 'Sat']
# tips.loc[행, 열]

tips.loc[0:1, ['total_bill', 'tip', 'day']]
```

## iloc

```python
tips.iloc[0:1, [0,1,4]]
```

```python
tips.loc[tips['day'] == 'Sat', ['total_bill', 'tip', 'day']]
```

```python
tips.loc[tips['total_bill'] <=11, :].head(1)
```

```python
tips. loc[tips['time'] == 'Dinner', :].head(1)
```

```python
#time이 Dinner이면서 total_bill이 11이하인 것만 조회
```

```python
result = tips.loc[tips['time'] == 'Dinner', :]
result.loc[result['total_bill'] <= 11, :]
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <=11), :]
# 조건식 2개
# tips.loc[(조건식 1) &(조건식 2), :]
```

```python
#iris
#품종이 virginica 이거나 sepal-lenght >= 5인  값만 가져오고
# 컬럼은 sepal_length, petal_length, species 만 가져오기
#loc 기반으로
```

```python
#iris.loc[(조건식1) or (조건식 2),['sepal_length', 'petal_length', 'species']]
#iris.loc[(조건식1) | (조건식 2),['sepal_length', 'petal_length', 'species']]
iris.loc[(iris['species'] == 'virginica') | (iris['sepal_length'] >= 5), ['sepal_length', 'petal_length', 'species']]
```
